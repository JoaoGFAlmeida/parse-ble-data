```python
import pandas as pd
import pandasql as ps
```

```python
df = pd.read_csv("2025-08-20_21-41-23.655.csv")
df.head()
```

```python
pd.set_option("display.max_colwidth", None)

query = """
SELECT 
    a.timestamp,
    a.mac   AS mac_a,
    a.rawData AS raw_a,
    a.rssi  AS rssi_a,
    b.mac   AS mac_b,
    b.rawData AS raw_b,
    b.rssi  AS rssi_b
FROM df a
JOIN df b 
    ON a.timestamp = b.timestamp 
   AND a.mac = 'C300001DA91C'
   AND b.mac = 'C300001DA912'
LIMIT 5;
"""

result = ps.sqldf(query, locals())
result.to_csv("raw_comparison.csv", index = False)

```
